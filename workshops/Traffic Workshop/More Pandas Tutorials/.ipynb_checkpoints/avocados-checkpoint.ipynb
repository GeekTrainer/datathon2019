```python
# Read a csv from a file.
df = pd.read_csv("avocado.csv")
df.head(3)
```

```python
# Drop single columns.
df = df.drop('Unnamed: 0', axis=1)
df.head(3)
```

```python
# Rename column using key-value pairs.
df = df.rename(columns={"AveragePrice": "Average Price"})
df.head(3)
```

```python
# Sort values of column in descending order.
df = df.sort_values('Average Price', ascending=False)
df.head(3)
```

```python
# Add multiple numerical columns together.
df['All Avocados'] = df['4046'] + df['4225'] + df['4770']
df.head(3)
```

```python
# Drop multiple columns.
df = df.drop('4046 4225 4770'.split(), axis=1)
df.head(3)
```

```python
# All these statements pull the same data out of the dataframe.
df['region'].head()
df.iloc[:, -1].head()
df.region.head()
```

```python
# Access 2nd value of 1st column (indexing from 0, so 1 over, 2 down).
df.iloc[2, 1]
```

```python
# Select all rows from one column.
df.loc[:, 'Date'].head()
```

```python
# Select multiple columns.
df[['Date','Average Price']].head()
```

```python
# Select single column.
df['Average Price'].head()
```

```python
# Filter dataframe with conditional.
specific_choice = df[df['Average Price'] > 0.5]
specific_choice.head()
```

```python
# Multiple conditionals, pay attention to the parentheses!
specific_choice = df[(df['Average Price'] > 0.5) & (df['year'] > 2017)]
specific_choice.head()
```

```python
# Filter by three conditionals.
exercise = df[(df['Average Price'] < 0.75) & (df['region'] == 'SanDiego') & (df['year'] == 2017)]
exercise.head()
```

```python
# List of unique places.
places = list(df['region'].unique())
places
```

```python
# Remove places that include given cities.
inclusive = ['WestTexNewMexico', 'California', 'West', 'Southeast', 'Northeast', 'Midsouth', \
             'Plains', 'TotalUS', 'NorthernNewEngland', 'SouthCentral']
single_region = [i for i in places if i not in inclusive]


by_region = df[df['region'].isin(single_region)]
len(by_region), len(df)
```

```python
# Proportion of total volumn taken up by San Diego (roughly).
sandiego = by_region[by_region['region'] == 'SanDiego']
sandiego_total = sandiego['Total Volume'].sum()
total = by_region['Total Volume'].sum()

print(sandiego_total, total)

prop_sandiego = sandiego_total / total
prop_sandiego
```

```python
# Filter using OR statement, proportion of total bags taken up by Atlanta and Denver (roughly).
atlanta_denver = by_region[(by_region['region'] == 'Atlanta') | (by_region['region'] == 'Denver')]
ad_total = atlanta_denver['Total Bags'].sum()
total = by_region['Total Bags'].sum()

prop_total_bags = ad_total / total
prop_total_bags
```

```python
by_region.head()
```

```python
# Group by a column and apply a method to the grouping.
by_region.groupby('region').mean().head()
```

```python
# Group by a column and apply a method to specific columns.
grouped = by_region.groupby('region')[['Total Volume', 'Total Bags']].sum()
grouped.head()
```

```python
regrouped = grouped.reset_index()
regrouped.head()
```

```python
regrouped = regrouped.set_index('Total Bags').sort_index()
regrouped['Proportion'] = regrouped.index / regrouped['Total Volume']
regrouped = regrouped.sort_values('Proportion')
regrouped.head()
```

```python
plt.figure(figsize=(14,10))
plt.yticks(fontsize=11)
plt.ylabel('country', fontsize=14)
plt.barh(regrouped.region, regrouped.Proportion, 1)
```
